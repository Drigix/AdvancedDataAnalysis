```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print(kmeans.predict([[10, 0]])[0])
```

```python
import pandas as pd
autos_pd = pd.read_parquet('autos.parquet', engine='pyarrow')
auto_mpg_pd = pd.read_parquet('auto-mpg.parquet', engine='pyarrow')
hungarian_heart_disease_pd = pd.read_parquet('hungarian-heart-disease.parquet', engine='pyarrow') 

autos_pd
```

```python
autos_pd_missing_values = autos_pd.isnull().sum()
autos_pd_total_cells = autos_pd.size
autos_pd_total_missing = autos_pd_missing_values.sum()
autos_pd_missing_ratio = autos_pd_total_missing / autos_pd_total_cells

auto_mpg_pd_missing_values = auto_mpg_pd.isnull().sum()
auto_mpg_pd_total_cells = auto_mpg_pd.size
auto_mpg_pd_total_missing = auto_mpg_pd_missing_values.sum()
auto_mpg_pd_missing_ratio = auto_mpg_pd_total_missing / auto_mpg_pd_total_cells

hungarian_heart_disease_pd_missing_values = hungarian_heart_disease_pd.isnull().sum()
hungarian_heart_disease_pd_total_cells = hungarian_heart_disease_pd.size
hungarian_heart_disease_pd_total_missing = hungarian_heart_disease_pd_missing_values.sum()
hungarian_heart_disease_pd_missing_ratio = hungarian_heart_disease_pd_total_missing / hungarian_heart_disease_pd_total_cells

data = {'row_count': [autos_pd.shape[0], auto_mpg_pd.shape[0], hungarian_heart_disease_pd.shape[0]],
        'col_count': [autos_pd.shape[1], auto_mpg_pd.shape[1], hungarian_heart_disease_pd.shape[1]],
        'missings_ratio': [autos_pd_missing_ratio, auto_mpg_pd_missing_ratio, hungarian_heart_disease_pd_missing_ratio],
        'class_count': [len(autos_pd['class'].unique()), len(auto_mpg_pd['class'].unique()), len(hungarian_heart_disease_pd['class'].unique())]}

df = pd.DataFrame(data)

print(df)
```
